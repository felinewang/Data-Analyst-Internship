```python
#part 1 

df = pd.DataFrame(columns = ['id','last_login','total_time'])
users = ends.device_id.unique()
for user in users:
    sample = ends[ends.device_id == user]
    last = sample.iloc[-1].create_time
    total = sample.session_duration.sum(axis = 0)
    df = df.append(pd.DataFrame({'id':[user],'last_login':[last],'total_time':[total]}),ignore_index=True)
    print(user)
```
